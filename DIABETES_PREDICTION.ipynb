Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data collection and analysis
PIMA Diabetes Dataset

```python
#loading the dataset to pandas dataset
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
pd.read_csv?
```

```python
#printing the first 5 rows of dataset
diabetes_dataset.head()
```

```python
#no of rows and columns in dataset
diabetes_dataset.shape
```

```python
#getting statistical measures of data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0 --> non diabetic
1 --> diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
#seperating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standarlization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarlized_data = scaler.transform(X)
```

```python
print(standarlized_data)
```

```python
X = standarlized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify =Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
#training svm classfier
classifier.fit(X_train, Y_train)
```

model evaluation

```python
#accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of training data: ', training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of test data: ', test_data_accuracy)
```

Making a predictive system

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

#changing input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape array as we r predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

#standarlize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0] == 0):
  print("The person is not diabetic")
else:
  print('The person is diabetic')
```

Saving the trained model
